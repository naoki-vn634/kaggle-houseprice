```python
import numpy as np
import pandas as pd
pd.set_option("display.max_rows",100)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("train.csv")
data.head()
```

```python
ans=pd.read_csv("test.csv")
ans.head()
```

```python
print(data.isnull().sum().head())
```

```python
print(ans.isnull().sum().head())
```

```python
data.dtypes.head()
```

```python
#欠損値の多い項目の削除
#Alleyの削除
data=data.drop("Alley",axis=1)
ans=ans.drop("Alley",axis=1)
#PoolQCの削除
data=data.drop("PoolQC",axis=1)
ans=ans.drop("PoolQC",axis=1)
#Fenceの削除
data=data.drop("Fence",axis=1)
ans=ans.drop("Fence",axis=1)
#MiscFeatureの削除
data=data.drop("MiscFeature",axis=1)
ans=ans.drop("MiscFeature",axis=1)
#FireplaceQuの削除
data=data.drop("FireplaceQu",axis=1)
ans=ans.drop("FireplaceQu",axis=1)
#種類の少ない項目の削除
#Utilitiesの削除
data=data.drop("Utilities",axis=1)
ans=ans.drop("Utilities",axis=1)
```

```python
#欠損値の補完
#BasementとGarageの欠損値は存在しない場合が考えられるためNoneでReplace
data["BsmtQual"].fillna('None', inplace=True)
ans["BsmtQual"].fillna('None', inplace=True)
data["BsmtCond"].fillna('None', inplace=True)
ans["BsmtCond"].fillna('None', inplace=True)
data["BsmtExposure"].fillna('None', inplace=True)
ans["BsmtExposure"].fillna('None', inplace=True)
data["BsmtFinType1"].fillna('None', inplace=True)
ans["BsmtFinType1"].fillna('None', inplace=True)
data["BsmtFinSF1"].fillna('None', inplace=True)
ans["BsmtFinSF1"].fillna('None', inplace=True)
data["BsmtFinType2"].fillna('None', inplace=True)
ans["BsmtFinType2"].fillna('None', inplace=True)
data["BsmtFinSF2"].fillna('None', inplace=True)
ans["BsmtFinSF2"].fillna('None', inplace=True)
data["BsmtUnfSF"].fillna('None', inplace=True)
ans["BsmtUnfSF"].fillna('None', inplace=True)
data["TotalBsmtSF"].fillna('None', inplace=True)
ans["TotalBsmtSF"].fillna(ans["TotalBsmtSF"].mean(), inplace=True)
data["GarageType"].fillna('None', inplace=True)
ans["GarageType"].fillna('None', inplace=True)
#GarageYrBltの欠損値を平均値補完
data["GarageYrBlt"].fillna(data["GarageYrBlt"].mean(), inplace=True)
ans["GarageYrBlt"].fillna(ans["GarageYrBlt"].mean(), inplace=True)
data["GarageFinish"].fillna('None', inplace=True)
ans["GarageFinish"].fillna('None', inplace=True)
data["GarageCars"].fillna('None', inplace=True)
ans["GarageCars"].fillna(ans["GarageCars"].mean(), inplace=True)
data["GarageArea"].fillna('None', inplace=True)
ans["GarageArea"].fillna(ans["GarageArea"].mean(), inplace=True)
data["GarageQual"].fillna('None', inplace=True)
ans["GarageQual"].fillna('None', inplace=True)
data["GarageCond"].fillna('None', inplace=True)
ans["GarageCond"].fillna('None', inplace=True)
data["MasVnrType"].fillna("None",inplace=True)
ans["MasVnrType"].fillna("None",inplace=True)
data["MasVnrArea"].fillna(0,inplace=True)
ans["MasVnrArea"].fillna(0,inplace=True)
ans["Exterior1st"].fillna('None',inplace=True)
ans["Exterior2nd"].fillna('None',inplace=True)
ans["BsmtFullBath"].fillna('None',inplace=True)
ans["BsmtHalfBath"].fillna('None',inplace=True)
ans["KitchenQual"].fillna('None',inplace=True)
ans["Functional"].fillna('None',inplace=True)
ans["SaleType"].fillna('None',inplace=True)
#Electricalは最頻要素で補完
data["Electrical"].fillna('SBrkr',inplace=True)
ans["MSZoning"].fillna('RL',inplace=True)
#Neighborhoodでグループし平均をとって補完
f = lambda x: x.fillna(x.mean())
data["LotFrontage"] = data.groupby("Neighborhood")["LotFrontage"].transform(f)
ans["LotFrontage"] = ans.groupby("Neighborhood")["LotFrontage"].transform(f)
```

```python
##FeatureEngineering
#総床面積を算出する
data["TotalFlrSF"]=data["1stFlrSF"]+data["2ndFlrSF"]
ans["TotalFlrSF"]=ans["1stFlrSF"]+ans["2ndFlrSF"]
#総ベランダ面積を算出する
data["TotalHousePorchSF"]=data["EnclosedPorch"]+data["OpenPorchSF"]+data["WoodDeckSF"]+data["3SsnPorch"]+data["ScreenPorch"]
ans["TotalHousePorchSF"]=ans["EnclosedPorch"]+ans["OpenPorchSF"]+ans["WoodDeckSF"]+ans["3SsnPorch"]+ans["ScreenPorch"]                               
#建設年から築年数への変換
data["Age"]=2018-data["YearBuilt"]
ans["Age"]=2018-ans["YearBuilt"]
#リフォームされた年から、されてからの期間への変換
data["YrSinceRemod"]=2018-data["YearRemodAdd"]
ans["YrSinceRemod"]=2018-ans["YearRemodAdd"]                             
#売年から売れてからの年への変換
data["YrSinceSold"]= 2018-data["YrSold"]
ans["YrSinceSold"]= 2018-ans["YrSold"]
#garageに関しても同じ処理を施す
data["GarageSinceYrBlt"]=2018-data["GarageYrBlt"]
ans["GarageSinceYrBlt"]=2018-ans["GarageYrBlt"]
```

```python
#階別床面積の項目の削除
data=data.drop("1stFlrSF",axis=1)
ans=ans.drop("1stFlrSF",axis=1)
data=data.drop("2ndFlrSF",axis=1)
ans=ans.drop("2ndFlrSF",axis=1)
#建設年の項目の削除
data=data.drop("YearBuilt",axis=1)
ans=ans.drop("YearBuilt",axis=1)
#GarageYrBltの項目の削除
data=data.drop("GarageYrBlt",axis=1)
ans=ans.drop("GarageYrBlt",axis=1)
```

```python
#欠損値の確認
print(data.isnull().any().head())
print("\n")
print(ans.isnull().any().head())
```

```python
#RabelEncoderでmapping
from sklearn.preprocessing import LabelEncoder

for i in range(data.shape[1]):
    if data.iloc[:,i].dtypes == object:
        lbl = LabelEncoder()
        lbl.fit(list(data.iloc[:,i].values) + list(ans.iloc[:,i].values))
        data.iloc[:,i] = lbl.transform(list(data.iloc[:,i].values))
        ans.iloc[:,i] = lbl.transform(list(ans.iloc[:,i].values))
```

```python
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```

```python
print('adjusted R^2')
#print('train score: %3f' % adjusted(lr.score(X_train_std, y_train_std), len(y_train_std), 2))
#print('test score : %3f' % adjusted(lr.score(X_test_std, y_test_std), len(y_test_std), 2))
print('train score: %3f' % adjusted(lr.score(X_train, y_train), len(y_train), 2))
print('test score : %3f' % adjusted(lr.score(X_test, y_test), len(y_test), 2))
```
